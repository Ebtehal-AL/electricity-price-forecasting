```python
import pandas as pd

hourly = pd.read_excel("../data/raw/hourly_original.xlsx", skiprows=9)
hourly.head()
```

```python
hourly = hourly[["Start date", "Germany/Luxembourg [€/MWh]"]]
hourly.head()
```

```python
hourly.columns = ["timestamp", "price"]
hourly.head()
```

```python
hourly["timestamp"] = pd.to_datetime(hourly["timestamp"], format="%b %d, %Y %I:%M %p", errors="coerce")
hourly["price"] = pd.to_numeric(hourly["price"], errors="coerce")
```

```python
hourly.info()
```

```python
hourly = hourly.dropna().reset_index(drop=True)
hourly.head()
```

```python
hourly.to_csv("../data/clean/hourly_clean.csv", index=False)
```
